```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris

data = load_iris()
data
```

```python
X = data.data          # (150, 4)
y = data.target        # 0, 1, 2
feature_names = data.feature_names
```

```python
feature_names
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 6))
for i, ax in enumerate(axes.flat):
    ax.hist(X[:, i], bins=20)
    ax.set_title(feature_names[i])
plt.tight_layout()
plt.show()
```

```python
feature_idx = 2  # petal length
data_by_class = [X[y == c, feature_idx] for c in np.unique(y)]

plt.boxplot(data_by_class, labels=data.target_names)
plt.title(feature_names[feature_idx])
plt.show()
```

```python
plt.scatter(X[:, 2], X[:, 3], c=y, cmap='viridis')
plt.xlabel(feature_names[2])
plt.ylabel(feature_names[3])
plt.show()
```

```python
corr = np.corrcoef(X.T)

fig, ax = plt.subplots()
im = ax.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
ax.set_xticks(range(4)); ax.set_xticklabels(feature_names, rotation=45, ha='right')
ax.set_yticks(range(4)); ax.set_yticklabels(feature_names)
plt.colorbar(im)
plt.title("Correlación entre features")
plt.show()
```
